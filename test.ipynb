```python
import sqlite3
```

```python
conn = sqlite3.connect('f1.db')
cur = conn.cursor()  #establish a connection to the database
```

```python
cur.execute("SELECT id,RaceName,DriverName,Round,status, Season from Results") #select data from the database using SQL syntax
```

```python
rows =  cur.fetchall()
for row in rows:
    print(row)
```

```python
conn.close() #close the connection
```
